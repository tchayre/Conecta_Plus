```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Read the Excel file
file_path = "/workspace/uploads/EMPRESAS_PALMAS_TO_1.xlsx"
df = pd.read_excel(file_path)

# Display basic information about the dataset
print("Dataset Overview:")
print(f"Number of rows: {df.shape[0]}")
print(f"Number of columns: {df.shape[1]}")
print("\nColumn Names:")
for col in df.columns:
    print(f"- {col}")

# Check data types and missing values
print("\nData Types and Missing Values:")
print(df.info())

# Count missing values per column
missing_values = df.isnull().sum()
print("\nMissing Values Count:")
print(missing_values[missing_values > 0])

# Display unique values for categorical columns
categorical_cols = ['Natureza Jurídica', 'Matriz ou Filial', 'Porte da Empresa', 
                    'Forma Tributação', 'Optante Simples', 'Optante MEI', 
                    'Situação Cadastral', 'UF', 'Município']

print("\nUnique Values in Categorical Columns:")
for col in categorical_cols:
    if col in df.columns:
        unique_values = df[col].unique()
        print(f"\n{col}: {len(unique_values)} unique values")
        if len(unique_values) < 20:  # Only print if there are fewer than 20 unique values
            print(unique_values)

# Analyze the distribution of business types
if 'Descrição CNAE Fiscal' in df.columns:
    print("\nTop 10 Business Types (CNAE Fiscal):")
    cnae_counts = df['Descrição CNAE Fiscal'].value_counts().head(10)
    print(cnae_counts)
    
    # Visualize the distribution of top business types
    plt.figure(figsize=(12, 6))
    cnae_counts.plot(kind='barh')
    plt.title('Top 10 Business Types in Palmas, TO')
    plt.xlabel('Count')
    plt.ylabel('Business Type')
    plt.tight_layout()
    plt.savefig('/workspace/business_types_distribution.png')
    plt.close()

# Analyze business size distribution
if 'Porte da Empresa' in df.columns:
    print("\nBusiness Size Distribution:")
    size_counts = df['Porte da Empresa'].value_counts()
    print(size_counts)
    
    # Visualize business size distribution
    plt.figure(figsize=(10, 6))
    size_counts.plot(kind='pie', autopct='%1.1f%%')
    plt.title('Business Size Distribution')
    plt.ylabel('')
    plt.tight_layout()
    plt.savefig('/workspace/business_size_distribution.png')
    plt.close()

# Analyze geographical distribution (by neighborhood)
if 'Bairro' in df.columns:
    print("\nTop 10 Neighborhoods:")
    bairro_counts = df['Bairro'].value_counts().head(10)
    print(bairro_counts)

# Display sample data
print("\nSample Data (First 5 rows):")
print(df.head())

# Summary of the data structure for the admin import functionality
print("\nData Structure Summary for Admin Import Functionality:")
print(f"The company data file contains {df.shape[0]} companies with {df.shape[1]} attributes per company.")
print("Key attributes that should be supported in the admin import functionality:")
important_cols = ['CNPJ', 'Razão Social', 'Nome Fantasia', 'Natureza Jurídica', 
                 'Porte da Empresa', 'CNAE Fiscal', 'Descrição CNAE Fiscal', 
                 'CNAEs Secundários', 'Logradouro', 'Número', 'Complemento', 
                 'Bairro', 'CEP', 'UF', 'Município', 'Telefone 1', 'Telefone 2', 'email']
for col in important_cols:
    if col in df.columns:
        missing_pct = df[col].isnull().mean() * 100
        print(f"- {col}: {missing_pct:.1f}% missing values")
```

```
from metagpt.tools.libs.terminal import Terminal
import json
import os

# Initialize terminal
terminal = Terminal()

# Function to create a simpler Python script for analysis
async def create_analysis_script():
    script = '''
import pandas as pd
import json
import sys
import os

try:
    # Read Excel file
    file_path = "/workspace/uploads/EMPRESAS_PALMAS_TO_1.xlsx"
    df = pd.read_excel(file_path)
    
    # Basic information
    result = {
        "num_rows": len(df),
        "num_columns": len(df.columns),
        "columns": list(df.columns),
        "data_types": {col: str(df[col].dtype) for col in df.columns},
        "missing_values": {col: int(df[col].isna().sum()) for col in df.columns},
        "sample_rows": df.head(5).to_dict(orient="records"),
        "unique_values": {}
    }
    
    # Get unique values for selected columns
    categorical_columns = ["Natureza Jurídica", "Matriz ou Filial", "Porte da Empresa", 
                          "Forma Tributação", "Optante Simples", "Optante MEI", 
                          "Situação Cadastral", "UF", "Município"]
    
    for col in categorical_columns:
        if col in df.columns:
            # Limit to 50 values maximum to avoid huge output
            unique_vals = df[col].dropna().unique().tolist()
            if len(unique_vals) <= 50:
                result["unique_values"][col] = unique_vals
            else:
                result["unique_values"][col] = unique_vals[:50] + ["... and more"]
    
    # Save results to JSON
    with open("/workspace/excel_analysis.json", "w", encoding="utf-8") as f:
        json.dump(result, f, ensure_ascii=False, indent=2)
    
    print("Analysis completed successfully!")
    
except Exception as e:
    error_result = {
        "error": str(e),
        "traceback": str(sys.exc_info())
    }
    with open("/workspace/excel_analysis_error.json", "w", encoding="utf-8") as f:
        json.dump(error_result, f, ensure_ascii=False, indent=2)
    print(f"Error analyzing file: {str(e)}")
'''
    
    script_path = "/workspace/analyze_excel.py"
    with open(script_path, "w") as f:
        f.write(script)
    
    return script_path

# Function to install necessary packages and run the analysis
async def analyze_excel_file():
    # Create the analysis script
    script_path = await create_analysis_script()
    
    # Install required packages
    print("Installing required packages...")
    await terminal.run("pip install pandas openpyxl")
    
    # Run the script
    print("Running Excel analysis script...")
    result = await terminal.run(f"python {script_path}")
    print(result)
    
    # Check if analysis was successful
    if os.path.exists("/workspace/excel_analysis.json"):
        # Read and display the analysis results
        with open("/workspace/excel_analysis.json", "r", encoding="utf-8") as f:
            analysis = json.load(f)
        
        print("\n=== Excel File Structure Analysis ===")
        print(f"Number of rows: {analysis['num_rows']}")
        print(f"Number of columns: {analysis['num_columns']}")
        
        print("\n=== Column Names ===")
        for col in analysis['columns']:
            print(f"- {col}")
        
        print("\n=== Data Types ===")
        for col, dtype in analysis['data_types'].items():
            print(f"- {col}: {dtype}")
        
        print("\n=== Missing Values ===")
        for col, count in analysis['missing_values'].items():
            if count > 0:
                print(f"- {col}: {count} missing values")
        
        print("\n=== Categorical Data Analysis ===")
        for col, values in analysis['unique_values'].items():
            print(f"\n{col} ({len(values)} unique values):")
            print(values)
        
        print("\n=== Sample Data (First 5 rows) ===")
        for i, row in enumerate(analysis['sample_rows'][:5]):
            print(f"\nRow {i+1}:")
            for key, value in row.items():
                # Print compact version for large fields
                if isinstance(value, str) and len(str(value)) > 50:
                    print(f"- {key}: {str(value)[:50]}...")
                else:
                    print(f"- {key}: {value}")
        
        return analysis
    else:
        print("Analysis failed. Check the error log.")
        if os.path.exists("/workspace/excel_analysis_error.json"):
            with open("/workspace/excel_analysis_error.json", "r") as f:
                error = json.load(f)
            print(f"Error: {error['error']}")
        return None

# Execute the analysis
await analyze_excel_file()
```
